```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

lena = cv2.imread('C:/Users/kaiyuchou/test_data/lena.bmp', cv2.IMREAD_GRAYSCALE)

x =  lena.shape[0] #
y =  lena.shape[1] #

lena_dilation = np.zeros(lena.shape, int)

#3-5-5-5-3 kernel
kernel = [[-2, -1], [-2, 0], [-2, 1],
          [-1, -2], [-1, -1], [-1, 0], [-1, 1], [-1, 2],
          [0, -2], [0, -1], [0, 0], [0, 1], [0, 2],
          [1, -2], [1, -1], [1, 0], [1, 1], [1, 2],
          [2, -1], [2, 0], [2, 1]]

for i in range(0, x, 1):
    for j in range(0, y, 1):
        if lena[i][j] > 0:
            max = 0
            for element in kernel:
                p, q = element
                if (i + p) >= 0 and (i + p) <= (x - 1) and \
                   (j + q) >= 0 and (j + q) <= (y - 1):
                    if lena[i + p][j + q] > max:
                        max = lena[i + p][j + q]
            for element in kernel:
                p, q = element
                if (i + p) >= 0 and (i + p) <= (x - 1) and \
                   (j + q) >= 0 and (j + q) <= (y - 1):
                    lena_dilation[i + p][j + q] = max
                                
cv2.imwrite('lena_gray_dilation.bmp', lena_dilation)
#pltshow cmap = 'gray'
plt.imshow(lena_dilation, cmap = 'gray')
```
